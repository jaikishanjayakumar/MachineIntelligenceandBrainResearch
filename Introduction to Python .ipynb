## Conditional Statements

In coding, when you want your software to perform different sets of operations depending on if a condition is true or false, you program it to check a condition. A condition is something that a computer can decide is either true or false.

Python supports the usual logical conditions from mathematics:

*   Equals: a == b
*   Not Equals: a != b
*   Less than: a < b
*   Less than or equal to: a <= b
*   Greater than: a > b
*   Greater than or equal to: a >= b

These conditions can be used in several ways, most commonly in "if statements" and loops (to tell the program when to break out of the loop).

### The **if** statement

An "if statement" is written by using the if keyword.

```python
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

In this example we use two variables, a and b, which are used as part of the if statement to test whether b is greater than a. As a is 33, and b is 200, we know that 200 is greater than 33, and so we print to screen that "b is greater than a".


### The **elif** statement


The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition".

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

In this example a is equal to b, so the first condition is not true, but the elif condition is true, so we print to screen that "a and b are equal".

### The **else** statement


The else keyword catches anything which isn't caught by the preceding conditions.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

In this example a is greater than b, so the first condition is not true, also the elif condition is not true, so we go to the else condition and print to screen that "a is greater than b".

## Loops (For and While)

Loops are a programming element that repeat a portion of code a set number of times until the desired process is complete. Repetitive tasks are common in programming, and loops are essential to save time and minimize errors.

If we want to print numbers from 1 to 10. We could either write the following inefficient piece of code






```python
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
```

**OR** 

we could make use of loops and achieve the same result using just 2 lines of code, as follows,


Python has two primitive loop commands:


*   for loops

### For Loops

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string). With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

### While Loops

With the while loop we can execute a set of statements as long as a condition is true.

```python
# Program to print i as long as i is less than 6

i = 1
while i < 6:
  print(i)
  i += 1
```

The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1. 

**Note:** remember to increment i, or else the loop will continue forever.

### Nested Loops

A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop".

Nested for loops can be used to iterate over the elements of 2D data structures.


```python
# A program to traverse every element of a two-dimensional array in Python

arr = [ [72, 85, 87, 90, 69], [80, 87, 65, 89, 85], [96, 91, 70, 78, 97], [90, 93, 91, 90, 94], [57, 89, 82, 69, 60] ]  

for x in arr:  # outer loop  
    for i in x:  # inner loop  
        print(i, end = " ") # print the elements; Here the argument 'end = " "' just makes sure we don't move to the next line (and we keep printing in the same line)
    print()  
```

### The **break** statement

With the break statement we can stop the loop before it has looped through all the items.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

In this examples, when the loop reaches the element "banana" and x is set to "banana", the condition `x == "banana"` gets satisfied which leads to the execution of the statement `break`, and the loop terminates right after it is executed. Thus, anything that comes after the element "banana" in the **fruits** list is not printed (processed). 

### The **continue** statement

With the continue statement we can stop the current iteration of the loop, and continue with the next iteration.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

After the continue statement is executed, the statements in the loop after the continue statement are skipped for the current iteration. Thus, `print(x)` is skipped for the iteration where `x == "banana"` is satisfied.

### Pass by value vs Pass by reference

In Python all functions work as pass by reference. This means that the changes made to a variable inside the function are also reflected in the calling function. Followig is an example illustrating this behaviour.



```python
def append_to_list(grocerylist, str):
  grocerylist.append(str)                     # the same object to which grocerylist is pointing to has been modified. 
  print("Value inside the function: ", grocerylist)
  return

# calling the append_to_list function
grocerylist = ["wheat", "sugar"]
print("value outside the function before calling the append_to_list: ", grocerylist)
append_to_list(grocerylist, "rice")
print("value outside the function after calling the append_to_list: ", grocerylist)

```

However, if we overwrite the argument inside the function then the reference to the local variable inside the function is reassigned. In this scenario, the change done to the variable inside the function will not be reflected outside the function. Following are few examples demonstrating this behaviour.


```python
# Example 1
def append_to_list(grocerylist, str):
  grocerylist = str                     # this assigns a new reference to grocerylist. Thus the value outside the function will not change 
  print("Value inside the function: ", grocerylist)
  return

# calling the append_to_list function
grocerylist = ["wheat", "sugar"]
print("value outside the function before calling the append_to_list: ", grocerylist)
append_to_list(grocerylist, "rice")
print("value outside the function after calling the append_to_list: ", grocerylist)

```

```python
# Example 2 
def increment(num):
  num = num + 1
  print("value inside the function", num)
  return                    # Note that the function does not return any value

# Calling increment function
num = 2
print("value outside the function before calling the function: ", num)
increment(num)
print("value outside the function after calling the function: ", num)
```

### Types of Function Arguments

There are four types of function arguments which are as follows:

* **Required**: The number of arguments should match the number of arguments in the function definition and the order of arguments must also be kept same.
* **Keyword**: When the calling function identifies the argument by its name during the function call, the argument is called as keyword argument. Here, the order is not important.
* **Default**: These are the arguments for which a default value is assigned during function definition. The default value is assumed only when the calling function does not provide a value for this variable.
* **Variable-length**: Sometimes you need a function that can take more number of arguments than what is specified at the time of defining the function. This can be achieved by defining a variable with *<variable_name>.

Following code provides an example of each of the argument type described above. 

```python
# Types of Function arguments

def print_details(name, age, gender):
  print("Name: ", name)
  print("Age: ", age)
  print("Gender: ", gender)
  return 

my_name =  "John Doe"
my_age = 21
my_gender = "M"

# calling using Required arguments
print("calling using Required arguments")
#print_details(name, age)             # this will generate an error as the number of arguments in function definition  and calling function are different.
#print_details(age, name, gender)     # this will execute but generate incorrect output as the order of arguments is not maintained.
print_details(my_name, my_age, my_gender)

```

```python
# calling using keyword arguments
print("calling using keyword arguments")
print_details(age=my_age, name=my_name, gender=my_gender)      # note that in this style order is not important but number of arguments should be same.
```

```python
print("Default arguments")
# default arguments are the ones that have a default value at the time of definition. The above function can be defined with a default gender variable as below.
def print_details(name, age, gender="Not specified"):
  print("Name: ", name)
  print("Age: ", age)
  print("Gender: ", gender)
  return

print_details(my_name, my_age)          # Note that the default value will be assumed as calling function does not provide a value for gender.
print_details(my_name, my_age, my_gender)
```

```python
# Variable Arguments
def print_details(name, age, gender, *moreinfo):
  print("Name: ", name)
  print("Age: ", age)
  print("Gender: ", gender)

  for info in moreinfo:
    print(info)
  return  

# calling the function with variable arguments
print("calling the function with variable arguments")
print_details(my_name, my_age, my_gender, "Address: Street XYZ, India", "University: IIT Madras")
```

## File I/O

Python provides built-in functions for working with files. You can manipulate files using **file** object. For modifying a file, you need to first open the file. For this, you can use Python's built-in open() function. This functions opens a file and returns a file object. 

Open function takes three arguments as shown below:

file_object = open(file_name, access_mode, buffering)

file_name: this argument is of string type. It should be set to the name of the file that you wish to open. The name should include the extension (for eg. .txt, .doc) as well.

access_mode: Python provides different modes in which the file can be opened. Some of the modes are read ('r'), write ('w'), read-write ('w+'), append (a) etc. The default value for this parameter is read mode.

Description of some of the common modes:

*   r: Opens a file for reading only and places the file pointer at beginning of the file.
*   r+: Opens a file for both reading and writing and the places the file pointer at the beginning of the file.
*   w: Opens a file for writing only. If the file does not exist, it creates a new file. If the file with the same name exists, the file is overwritten.
*   a: opens a file for appending and places the file pointer at the end of the file. If the file does not exist, it creates a new file for writing.

buffering: This argument specifies the buffer size to be used by the file. A buffer stores a chunk of data from the Operating System's file stream until it is consumed. This takes integer values. 
*   0: No buffering
*   1: buffers a line
*   positive value: bufferes the number of bytes specified
*   negative value: uses system's default buffer size. This is the **default** value.

### Reading and Writing to a file

For accessing files, the file object provides two methods, namely read() and write(). write() method writes a string to an open file. read() method reads a string from an open file. It takes an argument that specifies the number of bytes to be read from the file. 

Following is a small code snippet to write and read from a text file.








```python
# code to create and write to a text file
f_obj = open("/content/sample_data/sample.txt", 'w')  # open the file in write mode
content = "This is my first Python Program. \nPython is easy!"    
f_obj.write(content)
f_obj.close()                                      # close the file   

```

```python
# read from the newly create sample.txt file

f_obj = open("/content/sample_data/sample.txt", 'r')    # open file in read mode
str_read = f_obj.read(10)                               # read 10 bytes from the file. 1 chracter takes 1 byte. Thus this is same as reading 10 characters.      
print(str_read)
```

## Reading a sample CSV data
You will be using csv library to read and write a csv file in python. This library provides functions to read, write modify data in a csv file. Below is a small code snippet demonstrating the same.

```python
# importing the csv library
import csv

path = "/content/sample_data/california_housing_train.csv"      # path to the CSV file
file = open(path)
csvreader = csv.reader(file)                                    # use csv.reader to read csv file
heading = next(csvreader)                                       # next() method reads and return the current row and moves to the next row   
print(heading)
rows = []
for row in csvreader:
        rows.append(row)
print(rows[0])                                                  # print 0th row          
```

## Plotting in Python

You will be learning to use matplotlib library which is one of the most popular library in python for plotting and data visualization. You will be plotting two lines on the same plot. The examples below demostrates the functions used for plotting.


```python
# Example 1

import matplotlib.pyplot as plt

# x axis values
x = [1,2,3]
# corresponding y axis values
y = [3,6,9]
 
# plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
 
# giving a title to my graph
plt.title('My first graph!')
 
# function to show the plot
plt.show()
```

```python
# Example 2
# Plotting graph corresponding to a given equation

# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np
 
# setting the x - coordinates
x = np.arange(0, 2*(np.pi), 0.1)                          # samples points uniformly at an interval of 0.1 from (0, 2*pi)
# setting the corresponding y - coordinates
y = np.sin(x)
 
# poltting the points
plt.plot(x, y)
 
# function to show the plot
plt.show()
```
